```python
import numpy as np  # linear algebra
import pandas as pd  # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

from sklearn.datasets import make_moons

from sklearn.preprocessing import StandardScaler

from matplotlib import pyplot as plt

colors = ['blue', 'red']
vectorizer = np.vectorize(lambda x: colors[x % len(colors)])

features, true_labels = make_moons(n_samples=1000, noise=0.05, random_state=42)
FeaturesDF = pd.DataFrame(features, columns=['X', 'Y']).head()
```

```python
TargetDF = pd.DataFrame(true_labels, columns=['Class Label']).head()
```

```python
plt.scatter(features[:, 0], features[:, 1], c=vectorizer(true_labels))
```

```python
# Linear SVM
model = SVC(kernel='linear', C=1.0, random_state=42)
scores = cross_val_score(model, features, true_labels, cv=10)
print(scores)
model.fit(features, true_labels)
predictions = model.predict(features)
plt.scatter(features[:, 0], features[:, 1], c=vectorizer(predictions))
```

```python
# Quadratic SVM
model = SVC(kernel='poly', C=1.0, random_state=42)
scores = cross_val_score(model, features, true_labels, cv=10)
print(scores)
model.fit(features, true_labels)
predictions = model.predict(features)
plt.scatter(features[:, 0], features[:, 1], c=vectorizer(predictions))
```

```python
# Gaussian SVM
model = SVC(kernel='rbf', C=1.0, random_state=42)
scores = cross_val_score(model, features, true_labels, cv=10)
print(scores)
model.fit(features, true_labels)
predictions = model.predict(features)
plt.scatter(features[:, 0], features[:, 1], c=vectorizer(predictions))
```
