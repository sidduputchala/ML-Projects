```python
import numpy as np 
import pandas as pd
```

```python
data = pd.read_csv("StudentsPerformance.csv")
data.head(5)
```

```python
data["parental level of education"].value_counts()
```

```python
education_dict = {"some college":0,"some high school":0,"high school":1,"associate's degree":2,"bachelor's degree":3,"master's degree":4}
```

```python
data = data.replace({"parental level of education":education_dict})
data.head()
```

```python
data["lunch"].value_counts()
```

```python
data["lunch"] = data.apply(lambda x: 1 if x.lunch == "standard" else 0,axis=1)
data["lunch"].value_counts()
```

```python
data.head()
```

```python
data["gender"] = data.apply(lambda x: 1 if x.gender == "male" else 0,axis=1)
data["gender"].value_counts()
```

```python
data["test preparation course"] = data["test preparation course"].apply(lambda x : 0 if x == "none" else 1)
data["test preparation course"].value_counts()
```

```python
data1 = pd.get_dummies(data["race/ethnicity"])
data1.shape
```

```python
data2 = pd.concat([data,data1],axis='columns')
data2.shape
```

```python
data2 = data2.drop("race/ethnicity",axis=1)
```

**Model Training**

```python
#mark this column
X = data2.drop('test preparation course',axis='columns')
y = data2['test preparation course']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5)
X_train.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([ 
    keras.layers.Dense(11,input_dim=11 ,activation="relu"),  
    keras.layers.Dense(6,activation="relu"),
    keras.layers.Dense(1,activation="sigmoid")
])
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy', 
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=1000)
```

```python
model.evaluate(X_test, y_test)
```
